```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("end_full_12_25.csv")
```

```python
df.head()
```

```python
df.drop(columns=["Tier.y", "Squad.y"])

df = df.rename(columns={"Tier.x" : "Tier",
                        "Year.y" : "Year",
                        "Squad.x" : "Squad"})
```

```python
df.columns
```

```python
new_order = ['Year','League_id', 'Position', 'Club.x', 'Club.y','Played', 'W',
       'D', 'L', 'Goals', 'GD', 'Squad', 'Avg_Age', 'Foreigners',
       'Avg_Market_Value', 'Total_Market_Value', 'Area', 'Country', 'Id',
       'Competition', 'Tier']
df = df[new_order]
```

```python
df.head(15)
```

```python
df.info()
```

```python
def convert_currency(value):
    if isinstance(value, str):
        value = value.replace('€', '').strip()
        if 'k' in value:
            return float(value.replace('k', '')) * 1000
        elif 'm' in value:
            return float(value.replace('m', '')) * 1000000
        elif 'bn' in value:
            return float(value.replace('bn', '')) * 1000000000
        else:
            return float(value)
    return value
```

```python
df['Avg_Market_Value_Numeric'] = df['Avg_Market_Value'].apply(convert_currency)
df['Total_Market_Value_Numeric'] = df['Total_Market_Value'].apply(convert_currency)
```

```python
df.head(3)
```

```python
df['Points'] = df['W'] * 3 + df['D'] * 1 + df['L'] * 0
```

```python
df.head(2)
```

```python
new_order = ['Year','League_id', 'Position', 'Club.x', 'Club.y','Played', 'W',
       'D', 'L', 'Goals', 'GD', "Points", 'Squad', 'Avg_Age', 'Foreigners',
       'Avg_Market_Value', 'Total_Market_Value', "Avg_Market_Value_Numeric"	,"Total_Market_Value_Numeric",'Area', 'Country', 'Id',
       'Competition', 'Tier']
df = df[new_order]
```

```python
df.head(4)
```

```python
df["Points_Per_Match"] = np.where(df["Points"] != 0, df["Points"] / df["Played"], 0)
```

```python
df[['Goals_For', 'Goals_Against']] = df['Goals'].str.split(':', expand=True)
```

```python
df.head(4)
```

```python
df["Goals_For"] = pd.to_numeric(df["Goals_For"])
df["Goals_Against"] = pd.to_numeric(df["Goals_Against"])
```

```python
df["Goals_For_Per_Match"] = np.where(df["Goals_For"] != 0, df["Goals_For"] / df["Played"], 0)
```

```python
df["Goals_Against_Per_Match"] = np.where(df["Goals_Against"] != 0, df["Goals_Against"] / df["Played"], 0)
```

```python
new_order = ['Year', 'League_id', 'Position', 'Club.x', 'Club.y', 'Played', 'W', 'D',
       'L', 'Goals', 'GD', 'Points', 'Points_Per_Match', 'Goals_For', 'Goals_Against',
       'Goals_For_Per_Match', 'Goals_Against_Per_Match', 'Squad', 'Avg_Age', 'Foreigners',
       'Avg_Market_Value', 'Total_Market_Value', 'Avg_Market_Value_Numeric',
       'Total_Market_Value_Numeric', 'Area', 'Country', 'Id',
       'Competition', 'Tier']
df = df[new_order]
```
